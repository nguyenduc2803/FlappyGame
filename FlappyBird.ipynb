```python
    import pygame
from random import randint

# Khởi tạo Pygame
pygame.init()
screen = pygame.display.set_mode((400, 600))
pygame.display.set_caption('Chim Bay')
clock = pygame.time.Clock()

# Màu sắc
WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

# Tải hình ảnh
background_img = pygame.image.load('anhnen.png')
background_img = pygame.transform.scale(background_img, (400, 600))

bird_img = pygame.image.load('conchim.png')
bird_img = pygame.transform.scale(bird_img, (35, 35))

tube_img = pygame.image.load('ong1.png')
tube_dd_img = pygame.image.load('ong2.png')

# Biến khởi tạo
x_conchim = 50
y_conchim = 350
chim_vay_canh = 0
moi_lan_roi = 0.5
ong_di_chuyen = 2
score = 0
game_started = False  # Kiểm tra trò chơi đã bắt đầu hay chưa
pausing = False  # Kiểm tra trò chơi có bị tạm dừng hay không

# Phông chữ
font1 = pygame.font.SysFont('san', 40)
font = pygame.font.SysFont('san', 20)

# Thuộc tính ống
tube_width = 50
d_2tube = 150
tube1_x, tube2_x, tube3_x = 400, 600, 800
tube1_height, tube2_height, tube3_height = randint(100, 400), randint(100, 400), randint(100, 400)

tube1_pass, tube2_pass, tube3_pass = False, False, False

running = True
while running:
    clock.tick(60)
    screen.fill(WHITE)
    screen.blit(background_img, (0, 0))

    # Vẽ ống
    tube1_img = pygame.transform.scale(tube_img, (tube_width, tube1_height))
    tube1 = screen.blit(tube1_img, (tube1_x, 0))

    tube2_img = pygame.transform.scale(tube_img, (tube_width, tube2_height))
    tube2 = screen.blit(tube2_img, (tube2_x, 0))

    tube3_img = pygame.transform.scale(tube_img, (tube_width, tube3_height))
    tube3 = screen.blit(tube3_img, (tube3_x, 0))

    # Vẽ ống đối diện
    tube1_dd_img = pygame.transform.scale(tube_dd_img, (tube_width, 600 - tube1_height - d_2tube))
    tube1_dd = screen.blit(tube1_dd_img, (tube1_x, tube1_height + d_2tube))

    tube2_dd_img = pygame.transform.scale(tube_dd_img, (tube_width, 600 - tube2_height - d_2tube))
    tube2_dd = screen.blit(tube2_dd_img, (tube2_x, tube2_height + d_2tube))

    tube3_dd_img = pygame.transform.scale(tube_dd_img, (tube_width, 600 - tube3_height - d_2tube))
    tube3_dd = screen.blit(tube3_dd_img, (tube3_x, tube3_height + d_2tube))

    # Di chuyển ống sang trái
    if game_started:
        tube1_x -= ong_di_chuyen
        tube2_x -= ong_di_chuyen
        tube3_x -= ong_di_chuyen

    # Đặt lại ống khi ra khỏi màn hình
    if tube1_x < -tube_width:
        tube1_x = 550
        tube1_height = randint(100, 400)
        tube1_pass = False

    if tube2_x < -tube_width:
        tube2_x = 550
        tube2_height = randint(100, 400)
        tube2_pass = False

    if tube3_x < -tube_width:
        tube3_x = 550
        tube3_height = randint(100, 400)
        tube3_pass = False

    # Vẽ chim
    conchim = screen.blit(bird_img, (x_conchim, y_conchim))

    # Làm chim rơi nếu trò chơi đã bắt đầu
    if game_started:
        y_conchim += chim_vay_canh
        chim_vay_canh += moi_lan_roi

    # Hiển thị hướng dẫn nếu trò chơi chưa bắt đầu
    if not game_started:
        guide_txt = font1.render("press space to start", True, BLUE)
        screen.blit(guide_txt, (50, 250))

    # Thêm điểm khi chim đi qua ống
    if game_started:
        score_txt = font.render("Score " + str(score), True, RED)
        screen.blit(score_txt, (5, 5))

        if tube1_x + tube_width <= x_conchim and not tube1_pass:
            score += 1
            tube1_pass = True
        if tube2_x + tube_width <= x_conchim and not tube2_pass:
            score += 1
            tube2_pass = True
        if tube3_x + tube_width <= x_conchim and not tube3_pass:
            score += 1
            tube3_pass = True

    # Kết thúc trò chơi nếu chim chạm đất
    if y_conchim >= 565:
        ong_di_chuyen = 0
        chim_vay_canh = 0
        game_over_txt = font1.render("Game Over, Score: " + str(score), True, RED)
        screen.blit(game_over_txt, (50, 250))
        space_txt = font.render("press space to start", True, BLUE)
        screen.blit(space_txt, (90, 300))
        pausing = True

    # Kiểm tra va chạm với ống
    if game_started:
        tubes = [tube1, tube2, tube3, tube1_dd, tube2_dd, tube3_dd]
        for tube in tubes:
            if conchim.colliderect(tube):
                ong_di_chuyen = 0
                chim_vay_canh = 0
                game_over_txt = font1.render("Game Over, Score: " + str(score), True, RED)
                screen.blit(game_over_txt, (50, 250))
                space_txt = font.render("press space to start", True, BLUE)
                screen.blit(space_txt, (90, 300))
                pausing = True
                break

    # Xử lý sự kiện
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                if not game_started:
                    game_started = True
                elif pausing:
                    x_conchim, y_conchim = 50, 350
                    tube1_x, tube2_x, tube3_x = 400, 600, 800
                    score = 0
                    chim_vay_canh = 0
                    ong_di_chuyen = 2
                    pausing = False
                chim_vay_canh = -7

    pygame.display.flip()

pygame.quit()
```
